To first  make a histogram and Pie chart first import the modules needed to access the CSV file and to create a variable that would act as the Data Frame:

```python
import pandas as pd
import numpy as np


Data = pd.read_csv("olympics.csv", encoding = 'unicode_escape', engine = 'python')

Data
```

Fill in the empty space to Countries so that the coulmn has a label

```python
empVal = Data.isnull().values.any()

Data.fillna('Countries',inplace =True)
```

Remove the orginal headers and replaces them with the first row

```python
Data.columns = Data.iloc[0]
Data = Data[1:]
Data
```

```python
import matplotlib.pyplot as plt

Data['Combined total']
```

Converts the column Combined total from strings to integers so that it will be sorted by its numerical value vs alphabetically

```python
Data['Combined total'] = Data['Combined total'].astype(int,errors ='ignore')
SortCombTotal = Data.sort_values(by=['Combined total'], ascending = False)[1:]
SortCombTotal
```

```python
totalMedals = SortCombTotal['Combined total'].sum()
totalMedals
```

Shows the histogram of how many countries has won a set total of medals showing that the USA is the only Country above 1500

```python
plt.hist(SortCombTotal['Combined total'])
plt.show()
```

Creates a top ten variable

```python
TopTen = SortCombTotal.head(10)
TopTen
```

Pie Chart of The Top Ten countries who obtained medeals

```python
plt.pie(TopTen['Combined total'], labels = TopTen['Countries'])
plt.show()
```
