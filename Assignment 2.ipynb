Example of the code reading the  podcasts csv file. It did not display correctly as a table all the way across but does have do newline.

Unamed colums were renamed to Data 1, Data 2, and lastly downloads

display.max_rows will print all the rows instead othe first 5 and the last 5

display.max_columns will print all the columns

encoding = 'unicode_escape was the only way i found i could get the data printed.


```python

import pandas as pd

pd.set_option('display.max_rows', None,'display.max_columns', None)

x = pd.read_csv("podcasts.csv", encoding = 'unicode_escape', engine ='python')

x.rename( columns={"Unnamed: 6" : "Data 1"},inplace=True)
x.rename( columns={"Unnamed: 8" : "Data 2"},inplace=True)
x.rename( columns={"Unnamed: 9" : "Downloads"},inplace=True)

print(x)
```

sorted by number of downloads

```python
x_sort = x.sort_values(by='Downloads' , ascending = False)
print(x_sort)
```

Calls the dataframe x and then calls for the column named Downloads and finds all the ones with more than 1 million downloads.

```python
range_pods = x[x['Downloads'] > 1000000]

print(range_pods)
```

To find the all the ancient era pod casts and then sort

```python
# if the column Era == Ancient then it will be printed.
ancient_date = x[x['Era'] == 'Ancient']

#sorts the ancient era podcasts by the most downloads
ancient_pods_sorted= ancient_date.sort_values(by='Downloads', ascending = False)

print(ancient_pods_sorted)
```

```python
#iloc prints the information
print(ancient_pods_sorted.iloc[0])
```

```python
release_date = ancient_pods_sorted['Release date'].iloc[0]
Title =  ancient_pods_sorted['Title'].iloc[0]
print("The release date of the anicent pod cast with the most :\n", Title,
     '\nRelease Date: \n', release_date)
```
